```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

```python
data = {
    'player': ['Lewandowski', 'Mbappe', 'Messi', 'Haaland', 'De Bruyne'],
    'team': ['Barcelona', 'PSG', 'Inter Miami', 'Man City', 'Man City'],
    'position': ['Striker', 'Forward', 'Forward', 'Striker', 'Midfielder'],
    'goals': [25, 30, 22, 28, 15],
    'assists': [5, 10, 12, 6, 18],
    'yellow_cards': [2, 3, 1, 2, 4],
    'red_cards': [0, 0, 0, 1, 0]
}

df_raw = pd.DataFrame(data=data)
df_raw
```

```python
df = df_raw.copy()
df.info()
```

```python
for col in ['player', 'team', 'position',]:
    df[col] = df[col].astype('category')

df.info()
```

df.describe() - to metoda Pandas, która generuje podstawowe statystyki opisowe dla kolumn numerycznych w DataFrame.
count – liczba niepustych wartości w kolumnie
mean – średnia
std – odchylenie standardowe
min – najmniejsza wartość
25% – pierwszy kwartyl (25. percentyl)
50% – mediana (50. percentyl)
75% – trzeci kwartyl (75. percentyl)
max – największa wartość
